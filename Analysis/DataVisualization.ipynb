```python
import matplotlib.pyplot as plt
import numpy as np

# Dati per P1
labels = ['Min.', 'Max.', 'Avg.']
values_P1 = [2.7, 5.7, 4]
values_P1_costs = [78.30, 103.30, 81.80]
values_P1_resources = [29.53, 3.95, 3.26]

# Dati per P1'
values_P1_prime = [1.7, 3.4, 1.9]
values_P1_prime_costs = [49.20, 89.20, 55.20]
values_P1_prime_resources = [19.58, 1.97, 3.54]

# Larghezza degli istogrammi
bar_width = 0.35

# Creazione dei grafici
fig, axs = plt.subplots(1, 3, figsize=(15, 5))  # Modificato il numero di righe e colonne

# Grafico per la distribuzione dei tempi di ciclo
bar_positions = np.arange(len(labels))
axs[0].bar(bar_positions - bar_width / 2, values_P1, bar_width, color='#f3d6ce', label='P1', edgecolor='black')
axs[0].bar(bar_positions + bar_width / 2, values_P1_prime, bar_width, color='#d0e3cc', label="P1'", alpha=0.7, edgecolor='black')

# Aggiungi valori e unit di misura all'interno degli istogrammi
for i, v in enumerate(values_P1):
    axs[0].text(i - bar_width / 2, v + 0.1, f'{v} Hours', color='black', ha='center')

for i, v in enumerate(values_P1_prime):
    axs[0].text(i + bar_width / 2, v + 0.1, f'{v} Hours', color='black', ha='center')

axs[0].set_title('Cycle Time Distribution')
axs[0].set_xticks(bar_positions)
axs[0].set_xticklabels(labels)
axs[0].legend()
axs[0].set_ylabel('Hours')  # Aggiungi l'unit di misura all'asse y

# Grafico per la distribuzione dei costi
bar_positions = np.arange(len(labels))
axs[1].bar(bar_positions - bar_width / 2, values_P1_costs, bar_width, color='#f3d6ce', label='P1', edgecolor='black')
axs[1].bar(bar_positions + bar_width / 2, values_P1_prime_costs, bar_width, color='#d0e3cc', label="P1'", alpha=0.7, edgecolor='black')

# Aggiungi valori e unit di misura all'interno degli istogrammi
for i, v in enumerate(values_P1_costs):
    axs[1].text(i - bar_width / 2, v + 0.1, f'{v} ', color='black', ha='center')

for i, v in enumerate(values_P1_prime_costs):
    axs[1].text(i + bar_width / 2, v + 0.1, f'{v} ', color='black', ha='center')

axs[1].set_title('Cost Distribution')
axs[1].set_xticks(bar_positions)
axs[1].set_xticklabels(labels)
axs[1].legend()
axs[1].set_ylabel('')  # Aggiungi l'unit di misura all'asse y
axs[1].tick_params(axis='x', which='both', bottom=False)  # Nascondi le etichette sull'asse x

# Grafico per l'utilizzo delle risorse
worker_labels = ['Worker 1', 'Worker 2', 'Worker 3']  # Nuove etichette per le risorse
worker_positions = np.arange(len(worker_labels))
axs[2].bar(worker_positions - bar_width / 2, values_P1_resources, bar_width, color='#f3d6ce', label='P1', edgecolor='black', linewidth=1.2)
axs[2].bar(worker_positions + bar_width / 2, values_P1_prime_resources, bar_width, color='#d0e3cc', label="P1'", alpha=0.7, edgecolor='black',linewidth=1.2)

# Aggiungi valori e unit di misura all'interno degli istogrammi
for i, v in enumerate(values_P1_resources):
    axs[2].text(i - bar_width / 2, v + 0.1, f'{v}%', color='black', ha='center')

for i, v in enumerate(values_P1_prime_resources):
    axs[2].text(i + bar_width / 2, v + 0.1, f'{v}%', color='black', ha='center')

axs[2].set_title('Resource Utilization')
axs[2].set_xticks(worker_positions)
axs[2].set_xticklabels(worker_labels)
axs[2].legend()
axs[2].set_ylabel('% of Resource Utilization')  # Aggiungi l'unit di misura all'asse y

# Regola gli spazi tra i subplots
plt.tight_layout()
plt.savefig('SimulationBenchmark.pdf')
# Mostra i grafici
plt.show()
```
